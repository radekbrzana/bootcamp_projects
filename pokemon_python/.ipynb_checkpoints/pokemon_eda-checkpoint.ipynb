```python
#Pokemon Minior has different capture rate values depending on form (meteorite ot core).
#Minior has been duplicated to include both values in the analysis.

minior2 = pokedata[pokedata.capture_rate == '30 (Meteorite)255 (Core)']
minior2.capture_rate = 30
minior2.capture_rate
```

```python
pokedata.loc[pokedata.name == 'Minior', 'capture_rate'] = 255
```

```python
pokedata_cr = pd.concat([pokedata, minior2], ignore_index = True)
```

```python
pokedata_cr['capture_rate'] = pd.to_numeric(pokedata_cr['capture_rate'])
```

```python
pokedata_cr.info()
```

```python
pokedata_cr[pokedata_cr.name == 'Minior'].capture_rate
```

```python
#is there correlation between capture rate and weight, height or base total (sum of key pokemon stats)?

cr_cols = ['weight_kg', 'height_m', 'base_total']
cr_colors = ['#6189bb', '#5e9268', '#f1674a']


for col, color in zip(cr_cols, cr_colors):
    fig, ax = plt.subplots()
    sns.scatterplot(data = pokedata_cr, x = col, y ='capture_rate', color = color).invert_yaxis()
    ax.set_ylabel('capture_rate')
    ax.set_xlabel(col)
    ax.set_title('Relationship between capture_rate and {}. Correlation: {}'.format(col, round(pokedata_cr['capture_rate'].corr(pokedata_cr[col]),2)))
```

Conclusions: size of Pokemon has small influence on capture rate, its base stats are more significant.

```python
# replikuję każdy wiersz (każdy pokemon występuje teraz dwa razy)

pokedata_repl = pd.DataFrame(np.repeat(pokedata.values, 2, axis=0))

# przydzielam ponownie nazwy kolumn
    
pokedata_repl.columns = pokedata.columns
```

```python
# kilka pokemonów ma ten sam typ w type1 i type2, naprawiam to zastępując type2 NaNem
pokedata_repl.type2 = pd.np.where(pokedata_repl.type2 == pokedata_repl.type1, np.nan, pokedata_repl.type2) 

# w co drugim indeksie zastępuję wartość z type 1 na type2
pokedata_repl.type1 = pd.np.where(pokedata_repl.index % 2 == 0, pokedata_repl.type2, pokedata_repl.type1)

# tworzę nowy df bez kolumny type2, która już nie jest potrzebna
#pokedata4 = pokedata3[['abilities', 'attack', 'base_total', 'capture_rate', 'defense', 'experience_growth',
  #                   'height_m', 'hp','name', 'pokedex_number', 'sp_attack', 'sp_defense', 'speed', 'type1', 'weight_kg',
  #     'generation', 'is_legendary']]

# pozbywam się zduplikowanych pokemonów, które nie miały drugiego typu (a więc teraz mają NaN w pierwszym typie)

pokedata_repl = pokedata_repl[pokedata_repl['type1'].notna()]
```
